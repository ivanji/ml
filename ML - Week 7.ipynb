#### Support Vector Machines (SVMs)

It's a classification technique that provides the ideal split between available data. SVMs look for the widest possible margin that separate the hyperplane and the groups to classify. The closest points to the margin are called support vectors. (If support vectors were deleted/modified it'd create a move in the hyperplane).





##### C Parameter

In SVM we handle misclassification to decide how much to penalise missclassified points. Default C parameters is = 1

Low C parameter prioritises simplicity (Soft margin) and allows for misclassification. A higher C pritoritises making few mistakes (tends to overfit the training samples). 

##### Choice of Kernel (Similarity function)

No Kernel (Linear Kernel) - Standard linear classifier. e.g. 'Predict Y = 1' if $\Theta^T x > 0$

A good idea to use SVM without a kernel (or use logistic regression) when _n_ is large (relative to _m_).

Kernels solve for the parameter's $\Theta$ values

*Gaussian Kernel* (Most popular kernel).

if $\sigma^2 is large$ then you tend to have a higher bias, lower variance classifier. If small, then inverse if tends to be applicable.

Choose Gaussian Kernel when _n_ is small and _m_ is intermediate, use SVM with Gaussian kernel.

When using Gaussian kernel is sometimes required by the kernel library that you write the similarity function. 

Note: Do perform feature scaling before using the Gaussian Kernel, to avoid certain features to dominate because of their values.


##### Intuition of SVMs

With SVMs with a safety margin factor. The goal is to create a hyperplane that leaves the maximum margin for both classes. The margin is the distance between the hyperplane and each of the existing classes. 

If y = 1, we want $\Theta x > 1$ (not just > 0)
If y = 0, we want $\Theta x < -1$ (not just < 0)
