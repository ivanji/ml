# Training our Neural Network

### Training and Testing Sets

We separate our total samples between training and testing. After training a set, we test our model using the testing set. Whenever we can choose between a simpler model that does a similar job over an complex one, we should always choose the simpler one.

#### Underfitting (error due to bias)

Oversimplyfying the problem using a solution that is inadequate, and which will provide more errors in our data.
Example to remember: Think of it as not studying for a test


#### Overfitting (error due to variance)

Using an overcomplicated solution that won't generalise well - adding unnecessary extra complexities. Overly specific.
Example to remember: Think of it as memorizing every word in the book for a test.



We should err on the overfitting side BUT by implementing solutions that will help us prevent overfitting.



#### Early Stopping





It's not accurate to think that models that always provide the smaller errors are the best ones, as this normally represet overfitting.

Example: Which give smaller error for 1,1 and -1,-1: <br />
Prediction: $\hat{y} = \sigma(w_1x_1 + w_2x_2 + b) $<br />
Prediction: $\hat{y} = \sigma(1 + 1 + 0) \hat{y} = \sigma(-1 + -1 + 0) $ // just right
OR

Prediction: $\hat{y} = \sigma(10 + 10 + 0) \hat{y} = \sigma(-10 + -10 + 0) $ // Too accurate

When we apply sigmoid function to smaller values we get a nice slope to the gradient descent, in comparison to larger values where it gets more difficult to apply gradient descent, as values are either too high or too low.
