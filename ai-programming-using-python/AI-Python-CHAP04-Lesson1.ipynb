## Neural Network Refresher

### Perceptrons

A simple, single layer network, whose activation function is a step function. such as sigmoid or RELu (prediction is 1 or 0).

#### Linear Boundaries

$$2x_1 + x_2 + b$$ or
$$w_1x_1 + w_2x_2 + b = 0$$ OR abbreviated: $$Wx + b = 0 $$
where we're trying to predict y, which is a label: 0 or 1

Weights and biases are the learnable parameters of your model. The values of these parameters before learning starts are initialised randomly (this stops them all converging to a single value). Then when presented with data during training, they are adjusted towards values that have correct output.

In an applied example, where we're predicting acceptance at a university, we could use Grades and Test as input features and a random bias of -18, meaning the result would be: 2*Test + Grades - 18 

And the actual prediction would mean that if Score > 0: Accept, and < 0: Reject.


with Higher Dimensions (n):

$$w_1x_1 + w_2x_2 + w_nx_n + b = 0$$ OR abbreviated: $$Wx + b = 0 $$

#### Softmax

Softmax function is used -instead of sigmoid- on multi-classification problem, where the function helps us calculate the probabilities among all output units from the scores, which all total to 1. This will provide an indication what the highest probability is. 

The softmax can be used for any number of classes. It's also used for hundreds and thousands of classes, for example in object recognition problems where there are hundreds of different possible objects.

Sigmoid function is more apt for binary classifiers.

#### Maximum Likelihood

To measure probability we divide the number of events by the number of positive of outcomes. So we verify that the model has positively predicted the correct class.

Increasing the probability that our predictions are in fact correct, by using the product of all probabilities. Our goal is to increase the probability and this process is known as maximum likelihood. But instead of using products for our calculations of a large number of inputs, it's best to use the sum. See Cross-entropy.

#### Cross-entropy

Refers to the sum of all negative log of the probabilities in our model. It's used to determine how good or bad our model is. The lowers the cross-entropy the most accurate our model.
