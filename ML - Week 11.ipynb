#### Getting Lots of Data and Artificial Data

It's not always necessary to collect data from the real world, as some sample datasets can be artificially created. For instance in text recognition problems, text/fonts can be artificially created using multiple radom fonts already installed in your system and changing background/foreground colours, as well as introducing some warping to emulate many possible distinct variations, which could also be found in the real-world. 

Other samples include adding noise to audio files, etc.

Make sure you have a low bias classifer before expending the effor. (Plot learning curves). E.g. keep increasing the number of features/number of hidden units in neural network until you have a low bias classifier.

How much work would it be to get 10x as much data as we currently have?
- Artificial data synthesis
- Collect/label it yourself (e.g. if it takes around 10 seconds to label an extra example, it'd take approx. 3.5 days of 8 hrs work to come up with an extra 10,000 training examples).
- Crowd source (E.g. Amazon Mechanical Turk)
