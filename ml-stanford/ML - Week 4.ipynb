##### Non-linear Hypotheses (with Neural Networks)

On Non-linear classification problems, with large amount of features, algorithms  such as linear/logistic regression, won't perform efficiently. There are better alternatives, such as Neural Networks.

------

As an example, in computer vision, when a computer is trying to identify an object within an image, just as small as 50px x 50px image would result in a total of ~ 3,000,000 features.

Number of possible combinations of 2 from the list of the 2500 pixels (50x50px). It would be equal to the number of different pairs (xi,xj) where i and j range from 1 to 2500.

$n^2/2 = 3,125,000$ =~ 3 million approx.

Number of features = all possible combinations of xi and xj = n*n

But these will contain each pair twice. So we divide it by 2. (Think in terms of Permutations and Combinations)


The formula is $C(n, 2) = \dbinom{n}{2}=\dfrac{n(n-1)}{2}$  plus n (when i=j) or equivalently $\dfrac{n(n+1)}{2}$. So 2500*2501 / 2 = 3,126,250

This formula is called Binomial coefficient

##### Model Representation
##### Classification when x_1 and x_2 are binary Y $\in$ {0,1}

This is an example of OR function. 1 or 0 can be true for $X_1$ or $X_2$

| weight | $X_1$ | $X_2$ | Total |
|--------------------------- |
| -10 | 0  | 0  | -10 |
| -10 | 20 | 0  | 10 |
| -10 | 0  | 20 | -10 |
| -10 | 20 | 20 | 30


Example of AND.


| weight | $X_1$ | $X_2$ | Total |
|--------------------------- |
| -30 | 0  | 0  | -30 |
| -30 | 20 | 0  | -10 |
| -30 | 0  | 20 | -10 |
| -30 | 20 | 20 | 10

\begin{align*}& h_\Theta(x) = g(-30 + 20x_1 + 20x_2) \newline \newline & x_1 = 0 \ \ and \ \ x_2 = 0 \ \ then \ \ g(-30) \approx 0 \newline & x_1 = 0 \ \ and \ \ x_2 = 1 \ \ then \ \ g(-10) \approx 0 \newline & x_1 = 1 \ \ and \ \ x_2 = 0 \ \ then \ \ g(-10) \approx 0 \newline & x_1 = 1 \ \ and \ \ x_2 = 1 \ \ then \ \ g(10) \approx 1\end{align*}

Example of Negation

Add a large negative weight to $x_i$

\begin{align*}& h_\Theta(x) = g(10 + 20x_1)
\newline &
x_1 = 0 \ then \ g(10) \approx 1
\newline &
x_2 = 1 \ then \ g(-10) \approx 0
\end{align*}


Putting all examples together to create sample output $x_1 XNOR x_2$

\begin{align*}AND:\newline\Theta^{(1)} &=\begin{bmatrix}-30 & 20 & 20\end{bmatrix} \newline NOR:\newline\Theta^{(1)} &= \begin{bmatrix}10 & -20 & -20\end{bmatrix} \newline OR:\newline\Theta^{(1)} &= \begin{bmatrix}-10 & 20 & 20\end{bmatrix} \newline\end{align*}


\begin{align*}\begin{bmatrix}x_0 \newline x_1 \newline x_2\end{bmatrix} \rightarrow\begin{bmatrix}a_1^{(2)} \newline a_2^{(2)} \end{bmatrix} \rightarrow\begin{bmatrix}a^{(3)}\end{bmatrix} \rightarrow h_\Theta(x)\end{align*}

$\Theta^{(1)}$ =\begin{bmatrix}-30 & 20 & 20 \newline 10 & -20 & -20\end{bmatrix}

$\Theta^{(2)}$ =\begin{bmatrix}-10 & 20 & 20\end{bmatrix}

\begin{align*}& a^{(2)} = g(\Theta^{(1)} \cdot x) \newline& a^{(3)} = g(\Theta^{(2)} \cdot a^{(2)}) \newline& h_\Theta(x) = a^{(3)}\end{align*}
